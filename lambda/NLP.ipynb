```python
import nltk 
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import pandas as pd
nltk.download('averaged_perceptron_tagger')
```

```python
df = pd.read_csv("WomensClothing.csv", sep=",", index_col=0)
```

```python
df.head()
df['ReviewText']=df['ReviewText'].apply(str)
```

```python
#check if they are any missung values
for col in df.columns:
    print(col, df[col].isnull().sum())
```

```python
#view table with most imporant values
rws = df.loc[:, ['Age', 'Title', 'ReviewText', 'Rating' , 'RecommendedIND', 'DepartmentName', 'ClassName']]
rws.head()
```

```python
#only rows for recommended items
rws = rws[rws['RecommendedIND'] != 0]
rws.head()
```

```python
#only rows where rating is 5
rws = rws[rws.Rating == 5]
rws
```

```python
#delete rows for intimates and lounge wear
rws = rws[rws.ClassName != "Intimates"]
rws = rws[rws.ClassName != "Lounge"]
rws = rws[rws.ClassName != "Swim"]
rws = rws[rws.ClassName != "Chemises"]
rws = rws[rws.ClassName != "Trend"]
rws = rws[rws.ClassName != "Sleep"]
rws = rws[rws.ClassName != "Legwear"]
rws = rws[rws.DepartmentName != "Intimate"]
rws
```

```python
#drop rows with nan values
rws.dropna(inplace=True)
rws
```

```python
#delete columns rating and recommended 
rws = rws.loc[:, ['Age', 'Title', 'ReviewText','DepartmentName' ,'ClassName']]
rws
```

```python
#get adjectives by using pos tagger
def get_adjectives(row):
    text = word_tokenize(row["ReviewText"])
    tags = pos_tag(text)
    adjectives = ""
    previous = ""
    for tag in tags:
        if tag[1] == "JJ":
            adjectives = previous + tag[0] + ", "
            previous = adjectives
    return adjectives
```

```python
#get nouns by using pos tagger
def get_nouns(row):
    text = word_tokenize(row["ReviewText"])
    tags = pos_tag(text)
    adjectives = ""
    previous = ""
    for tag in tags:
        if tag[1] == "NN":
            adjectives = previous + tag[0] + ", "
            previous = adjectives
    return adjectives
```

```python
#replace fine gauge with thin sweater
def replace_fine(row):
    if row["ClassName"] == "Fine gauge":
        return "thin sweater"
    elif row["ClassName"] == "Outerwear":
        return "warm jacket"
    elif row["ClassName"] == "Blouses":
        return "blouse"
    elif row["ClassName"] == "Knits":
        return "knit sweater"
    elif row["ClassName"] == "Dresses":
        return "dress"
    elif row["ClassName"] == "Jackets":
        return "jacket"
    elif row["ClassName"] == "Skirts":
        return "skirt"
    elif row["ClassName"] == "Sweaters":
        return "sweater"
    elif row["ClassName"] == "Pants":
        return "pants"
    elif row["ClassName"] == "Jeans":
        return "jeans"
    elif row["ClassName"] == "Shorts":
        return "shorts"
    else:
        return row["ClassName"]
```

```python
#add adjectives to column adjectives and replace fine gauge
rws['Adjectives'] = rws.apply(get_adjectives, axis=1)
rws['ClassName'] = rws.apply(replace_fine, axis=1)
rws.head()
```

```python
#add nouns to column nouns
rws['Nouns'] = rws.apply(get_nouns, axis=1)
rws.head()
```

```python
#get events by filtering nouns
def get_events(row):
    events = ""
    if ("party" in row["ReviewText"]):
        events += " party"
    if ("going out" in row["ReviewText"]):
        events += " going out"
    if ("bar" in row["Nouns"]):
        events += " bar"
    if ("drinks" in row["Nouns"]):
        events += " drinks"
    if "business" in row["Nouns"]:
        events += " business"
    if "work" in row["Nouns"]:
        events += " work"
    if "dinner" in row["Nouns"]:
        events += " dinner"
    if "date" in row["Nouns"]:
        events += " date"
    if "vacation" in row["Nouns"]:
        events += " vacation"
    if "holiday" in row["Nouns"]:
        events += " holiday"
    return events
    
```

```python
#get season by filtering nouns
def get_seasons(row):
    seasons = ""
    if "summer" in row["Nouns"]:
        seasons += "summer"
    if "spring" in row["Nouns"]:
        seasons += " spring"
    if "winter" in row["Nouns"]:
        seasons += " winter"
    if "fall" in row["Nouns"]:
        seasons += "fall"
    if "autumn" in row["Nouns"]:
        seasons += " autumn"
    return seasons
```

```python
#add events and season to columns
#rws['Events'] = rws.apply(get_events, axis=1)
rws['Seasons'] = rws.apply(get_seasons, axis=1)
```
